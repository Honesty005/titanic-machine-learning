1Ô∏è‚É£ Import libraries

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

#Why:
#pandas ‚Üí data handling
#RandomForestClassifier ‚Üí machine learning model
```

2Ô∏è‚É£ Load the data

```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
train_data.head()

#Key idea:
#train_data has the answer (Survived)
#test_data does not ‚Äî that‚Äôs what we predict
```

3Ô∏è‚É£ Quick exploratory analysis

```python
women = train_data.loc[train_data["Sex"] == "female", "Survived"]
men = train_data.loc[train_data["Sex"] == "male", "Survived"]

rate_women = women.mean()
rate_men = men.mean()

rate_women, rate_men

#Interpretation (VERY IMPORTANT):
#Women survived at a much higher rate than men ‚Üí gender is a strong predictor.
```

4Ô∏è‚É£ Select features

```python
features = ["Pclass", "Sex", "SibSp", "Parch"]

#Why these?
#Pclass ‚Üí social status
#Sex ‚Üí strong survival signal
#SibSp, Parch ‚Üí family size
```

5Ô∏è‚É£ Prepare training data

```python
X = pd.get_dummies(train_data[features])
y = train_data["Survived"]

# Convert categorical variables (Sex) into numbers.
# The pandas.get_dummies() function converts categorical data 
# (text labels like "Red," "Green," "Blue") into dummy variables (binary 0s and 1s).
# This process is known as one-hot encoding.
```

6Ô∏è‚É£ Prepare test data

```python
X_test = pd.get_dummies(test_data[features])

# Ensure train and test have same columns
# Align train and test datasets to ensure they have the same feature columns
X, X_test = X.align(X_test, join="left", axis=1, fill_value=0)
```

7Ô∏è‚É£ Train the model

```python
model = RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    random_state=1
)

model.fit(X, y)

# We train many decision trees
# Each tree votes
# Final result = majority vote
```

8Ô∏è‚É£ Make predictions

```python
predictions = model.predict(X_test)
# Now the model is guessing who survived.
```

9Ô∏è‚É£ Create submission file

```python
submission = pd.DataFrame({
    "PassengerId": test_data["PassengerId"],
    "Survived": predictions
})

submission.head()
```

üîü Save submission

```python
submission.to_csv("submission.csv", index=False)
# Create submission file for Kaggle
```
